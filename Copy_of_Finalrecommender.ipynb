```python
def convert_to_3(obj):
  L = []
  count = 0
  for i in ast.literal_eval(obj):
    if count != 3:
       L.append(i['name'])
       count += 1
    else:
      break
  return L
```

```python
movies1['cast'] = movies1['cast'].apply(convert_to_3)
```

```python
# in crew only job = director is important
```

```python
def retrive_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies1['crew'] = movies1['crew'].apply(retrive_director)
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  
  return " ".join(y)
```

```python
new_dataframe['tags'] = new_dataframe['tags'].apply(stem)
```

```python
# on the basis of tags we need to show similarity
# similarity directly proportional to text similarity
# now how to compare text one way similar words butits not good way
# one way is to convert text in vector
# closest vectors can be recommended
# technique bag of words is technique to convert to vectors
# bag of words is concatenete all words and finding most commen words 
# then in each movie how many times word is comming
# in sklearn we have direct library
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_dataframe['tags']).toarray()
```

```python
# now every movie has each vector
# now we nedd to compare each movie with each other movie
# but euclidien distance is not good in higher value
# now we can compare with cosine simialrity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(vectors)
```

```python
similarity_matrix
```

```python
df1 = pd.DataFrame(similarity_matrix,index=similarity_matrix[:,0])
```

```python
def recommend(movie):
  movieindex = new_dataframe[new_dataframe['title'] == movie].index[0]
  distances = similarity_matrix[movieindex]
  movie_list =  sorted(list(enumerate(distances)),reverse = True,key=lambda x:x[1])[1:10]
  for i in movie_list:
    print(new_dataframe.iloc[i[0]].title)
  return
```
